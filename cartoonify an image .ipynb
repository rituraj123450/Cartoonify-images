```python
import cv2 
import numpy as np 


img=cv2.imread("WIN_20201108_16_01_05_Pro.jpg")
```

```python
from matplotlib import pyplot as plt
plt.imshow(img)
plt.show()
```

```python
image=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.figure(figsize=(10,10))
plt.imshow(image)
plt.show()
```

```python
img_small=cv2.pyrDown(image)
```

```python
num_iter=5
for _ in range(num_iter):
    img_small=cv2.bilateralFilter(img_small,d=9,sigmaColor=9,sigmaSpace=7)
```

```python
img_rgb=cv2.pyrUp(img_small)
```

```python
plt.imshow(img_rgb)
plt.show()
```

```python
img_gray=cv2.cvtColor(img_rgb,cv2.COLOR_RGB2GRAY)
img_blur=cv2.medianBlur(img_gray,7)
```

```python
img_edge=cv2.adaptiveThreshold(img_blur,255,cv2.ADAPTIVE_THRESH_GAUSSIAN_C,cv2.THRESH_BINARY,7,2)
```

```python
plt.imshow(img_edge)
plt.show()
```

```python
img_edge=cv2.cvtColor(img_edge,cv2.COLOR_GRAY2RGB)
```

```python
plt.imshow(img_edge)
plt.show()
```

```python
array=cv2.bitwise_and(image,img_edge)
```

```python
plt.figure(figsize=(10,10))
plt.imshow(array)
plt.show()
```
